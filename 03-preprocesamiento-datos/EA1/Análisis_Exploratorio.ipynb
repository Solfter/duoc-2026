```python
# ==========================================================
# CELDA 1: IMPORTAR LIBRERÍAS Y CARGAR EL ARCHIVO
# ==========================================================
import pandas as pd
from google.colab import files

uploaded = files.upload()
archivo = next(iter(uploaded))

df = pd.read_csv(archivo)

print("Dimensiones del dataset:")
print(df.shape)

print("\nPrimeras filas:")
display(df.head())
```

```python
# ==========================================================
# CELDA 2: LIMPIAR Y RENOMBRAR COLUMNAS
# ==========================================================
df.columns = df.columns.str.strip()

df = df.rename(columns={
    "S.NO": "s_no",
    "Name": "name",
    "Nationality": "nationality",
    "Current Rank": "current_rank",
    "Previous Year Rank": "previous_year_rank",
    "Sport": "sport",
    "Year": "year",
    "earnings ($ million)": "earnings"
})

# Limpiar la columna sport para evitar diferencias por mayúsculas o espacios
df["sport"] = df["sport"].astype(str).str.strip().str.title()

print("Columnas del dataframe:")
print(df.columns)
```

```python
# ==========================================================
# CELDA 3: IDENTIFICAR EL TIPO DE DATOS DE CADA VARIABLE
# ==========================================================
print("Tipos de datos:")
print(df.dtypes)
```

```python
# ==========================================================
# CELDA 4: ESTADÍSTICAS BÁSICAS CON DESCRIBE
# ==========================================================
print("Estadísticas básicas de las variables numéricas:")
display(df[["current_rank", "year", "earnings"]].describe())

print("Estadísticas básicas de las variables categóricas:")
display(df[["name", "nationality", "previous_year_rank", "sport"]].describe())
```

```python
# ==========================================================
# CELDA 5: MODA DE VARIABLES CATEGÓRICAS
# ==========================================================
print("Moda de la variable nationality:", df["nationality"].mode()[0])
print("Moda de la variable previous_year_rank:", df["previous_year_rank"].mode()[0])
print("Moda de la variable sport:", df["sport"].mode()[0])

# En name puede haber empate o poca utilidad, pero igualmente se puede mostrar
print("Moda de la variable name:", df["name"].mode()[0])
```

```python
# ==========================================================
# CELDA 6: TRANSFORMAR 'earnings' EN VARIABLE CATEGÓRICA
# ==========================================================
# Se crean 5 intervalos de clase
df["earnings_class"] = pd.cut(df["earnings"], bins=5)

print("Ejemplo de la nueva variable categórica:")
display(df[["earnings", "earnings_class"]].head(15))
```

```python
# ==========================================================
# CELDA 7: FRECUENCIA DE CADA CLASE
# ==========================================================
# Frecuencia absoluta: cantidad de datos en cada intervalo
frecuencia_absoluta = df["earnings_class"].value_counts().sort_index()

# Frecuencia relativa: proporción respecto del total
frecuencia_relativa = df["earnings_class"].value_counts(normalize=True).sort_index()

# Porcentaje
porcentaje = frecuencia_relativa * 100

tabla_frecuencias = pd.DataFrame({
    "Frecuencia absoluta": frecuencia_absoluta,
    "Frecuencia relativa": frecuencia_relativa,
    "Porcentaje": porcentaje
})

print("Tabla de frecuencias de la variable earnings categorizada:")
display(tabla_frecuencias)
```

```python
# ==========================================================
# CELDA 8: RESPUESTA A LA PREGUNTA A
# ¿Quién es la persona que más dinero ganó en un año?
# ==========================================================
fila_max = df.loc[df["earnings"].idxmax()]

print("a) ¿Quién es la persona que más dinero ganó en un año?")
print("Nombre:", fila_max["name"])
print("Año:", fila_max["year"])
print("Deporte:", fila_max["sport"])
print("Ganancia:", fila_max["earnings"], "millones de dólares")
```

```python
# ==========================================================
# CELDA 9: RESPUESTA A LA PREGUNTA B
# ¿Cuál es el deporte que más se repite dentro de la lista?
# ==========================================================
deporte_mas_repetido = df["sport"].mode()[0]
cantidad_deporte = df["sport"].value_counts()[deporte_mas_repetido]

print("b) ¿Cuál es el deporte que más se repite dentro de la lista?")
print("Deporte:", deporte_mas_repetido)
print("Cantidad de veces que aparece:", cantidad_deporte)
```

```python
# ==========================================================
# CELDA 10: RESPUESTA A LA PREGUNTA C
# ¿Cuál es el deporte donde más dinero se gana?
# ==========================================================
ganancias_por_deporte = df.groupby("sport")["earnings"].sum().sort_values(ascending=False)

print("Ganancia total por deporte:")
display(ganancias_por_deporte)

print("c) ¿Cuál es el deporte donde más dinero se gana?")
print("Respuesta:", ganancias_por_deporte.idxmax())
print("Ganancia total:", ganancias_por_deporte.max(), "millones de dólares")
```

```python
# ==========================================================
# CELDA 11: RESPUESTA A LA PREGUNTA D
# Si quisiera apostar a un jugador de Basketball según su ranking
# ==========================================================
basketball = df[df["sport"] == "Basketball"]
basketball = basketball.sort_values(by="current_rank")

mejor_basket = basketball.iloc[0]

print("d) Si quisiera apostar a un jugador de Basketball según su ranking, ¿a cuál debería apostar?")
print("Nombre:", mejor_basket["name"])
print("Ranking actual:", mejor_basket["current_rank"])
print("Año:", mejor_basket["year"])
print("Ganancia:", mejor_basket["earnings"], "millones de dólares")
```
