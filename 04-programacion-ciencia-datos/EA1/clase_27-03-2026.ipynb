```python
import pandas as pd

df = pd.DataFrame({
    'Genero': ['M', 'F', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M'],
    'Carrera': ['Ing. Datos', 'Ing. Comercial', 'Ing. Datos', 'Administración',
                'Ing. Comercial', 'Administración', 'Ing. Datos', 'Ing. Comercial',
                'Administración', 'Ing. Datos', 'Ing. Comercial', 'Administración'],
    'Ciudad': ['Santiago', 'Temuco', 'Santiago', 'Concepción',
               'Temuco', 'Santiago', 'Concepción', 'Temuco',
               'Santiago', 'Concepción', 'Temuco', 'Santiago'],
    'Edad': [20, 22, 21, 24, 23, 25, 20, 22, 24, 21, 23, 26],
    'Horas_Estudio': [10, 8, 12, 6, 9, 7, 11, 8, 6, 13, 9, 5],
    'Nota_Matematicas': [5.8, 5.4, 6.1, 4.9, 5.6, 5.0, 6.0, 5.5, 5.1, 6.3, 5.7, 4.8],
    'Nota_Programacion': [6.0, 5.6, 6.4, 5.1, 5.8, 5.2, 6.1, 5.7, 5.0, 6.5, 5.9, 4.9]
})

df
```

## agregar una columna promedio

```python
df['Promedio_Notas'] = (df['Nota_Matematicas'] + df['Nota_Programacion'] ) / 2

df
```

## agregar columna clasificacion de notas

```python
def clasificar_rendimiento(promedio):
    if promedio < 5.4:
        return 'Bajo'
    elif promedio < 6.0:
        return 'Medio'
    else:
        return 'Alto'

df['Rendimiento'] = df['Promedio_Notas'].apply(clasificar_rendimiento)

df
        
```

## convertir las columnas categoricas a columnas (0/1) get_dummies

```python
df_codificado = pd.get_dummies(
        df,
        columns=['Genero', 'Carrera', 'Ciudad', 'Rendimiento'],
        dtype=int    
)

df_codificado
```

```python
print('Columnas originales: ', df.columns.tolist())
print('Cantidad de columnas antes de codificar: ', len(df.columns))
print('Cantidad de columnas despues de codificar: ', len(df_codificado.columns))
```
