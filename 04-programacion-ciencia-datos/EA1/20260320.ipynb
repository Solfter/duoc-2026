### Metodos en DataFrames para limpieza

* isnull(): Devuelve True o False si el valor es null
* duplicated(): Devuelve True o False si el valor esta duplicado
* drop_duplicated(): Elimina los registros duplicados

```python
import pandas as pd

df = pd.DataFrame({
    "Nombre": ["Juan", "Maria", "Juan", None],
    "Edad": [20, None, 20, 22],
    "Nota": [5.0, 6.5, 5.0, None]
})
display(df)

df['Edad'] = df['Edad'].fillna(df['Edad'].mean())
print("="*50, "\nNulos de Edad como Promedio de Edad")
display(df)

df['Nota'] = df['Nota'].fillna(df['Nota'].mean())
print("="*50, "\nNulos de Nota como Promedio de Nota")
display(df)

df = df.dropna()
print("="*50, "\nEliminación de Nulos")
display(df)

df = df.drop_duplicates()
print("="*50, "\nEliminacion de duplicados")
display(df)

```
