### Problema 3 💻/🐍

Una startup de inteligencia artificial lanza una plataforma web que usa modelos de lenguaje.

Durante los 20 primeros días, el número de usuarios registrados se puede aproximar por la función:

$$U(t)=500(1-e^{-0,1t})$$

Donde $t$ corresponde al tiempo en días desde el lanzamiento.

a) Identifique las variables dependiente e independiente.

* Var. Indep.: Tiempo en días desde el lanzamiento
* Var. Dep: Número de usuarios registrados

b) Indique el dominio contextualizado de la función.

* Dom U(t): [0, 20]

c) Genere e imprima una tabla que muestre la cantidad de usuarios registrados durante cada uno de los 10 primeros días desde el lanzamiento de la plataforma web.

````python
t = sp.Symbol('t')
f = sp.Lambda(t, 500 * (1 - sp.exp(-0.1 * t)))

dom = [n for n in range(21)]
resultado = [round(f(n), 2) for n in dom]

df = pd.DataFrame({
    't': dom,
    'U(t)': resultado
})    

df
````

d) ¿Durante qué día se alcanzan los 400 usuarios?

````python
ecuacion = sp.Eq(500 * (1 - sp.exp(-0.1 * t)), 400)
solucion = sp.solve(ecuacion, t)[0].round()
print(f'Se alcazaron los 400 usuarios al dia {solucion}')
````

e) Grafique la función $U(t)$ considerando el dominio contextualizado.

````python
plt.plot(df['t'], df['U(t)'])
plt.scatter(df['t'], df['U(t)'])
plt.title('Usuarios diarios de la aplicación')
plt.xlabel('Días')
plt.ylabel('Número de usuarios')
plt.grid()
plt.show()
````

````python
#@title Gráfico interactivo de función lineal
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact


def lineal(m=1, n=0):
    x = np.linspace(-10, 10, 200)
    y = m*x + n

    plt.figure(figsize=(6,6))
    plt.plot(x, y, label=f'f(x) = {m:.1f}x + {n:.1f}')

    # Ejes principales
    plt.axhline(0, color='gray', lw=1)
    plt.axvline(0, color='gray', lw=1)

    # Punto de intersección con eje y
    plt.scatter(0, n, color='red', s=60, zorder=5, label='Intersección con eje y')

    # Configuración de ejes y grilla
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.xticks(np.arange(-10, 11, 1))
    plt.yticks(np.arange(-10, 11, 1))
    plt.grid(True, which='both', linestyle='--', linewidth=0.7)

    plt.legend()
    plt.show()
    return None

# Deslizadores interactivos
_ = interact(lineal, m=(-5,5,0.1), n=(-5,5,0.1))
````

### Problema 11 💻/🐍



A continuación, se presenta un gráfico interactivo de la función $ReLU(x)$ paramétrica dada por:

$$f(x)=a\cdot ReLU(x-b)+c$$

A partir del análisis de esta gráfica, responda las siguientes preguntas.

````python
#@title ReLU paramétrica
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as W
from IPython.display import clear_output, display

x = np.linspace(-6, 6, 400)

# Deslizadores
a = W.FloatSlider(value=1.0, min=-3, max=3, step=0.1, description='a')
b = W.FloatSlider(value=0.0, min=-3, max=3, step=0.1, description='b')
c = W.FloatSlider(value=0.0, min=-3, max=3, step=0.1, description='c')

out = W.Output()

def dibujo(*_):
    with out:
        clear_output(wait=True)
        y = a.value * np.maximum(0, x - b.value) + c.value
        fig, ax = plt.subplots(figsize=(7,4))
        ax.plot(x, y, lw=2, label='f(x)')
        ax.axvline(b.value, color='purple', ls=':', lw=1.5, label='umbral b')
        ax.scatter([b.value],[c.value], s=70, color='red', zorder=5, label='bisagra (b,c)')
        y0 = a.value*np.maximum(0, -b.value) + c.value
        ax.scatter([0],[y0], s=60, color='black', zorder=5, label='intersección y')
        ax.axhline(0, color='gray', lw=1); ax.axvline(0, color='gray', lw=1)
        ax.set_xlim(-6,6); ax.set_ylim(-6,10)
        ax.set_xticks(np.arange(-6,7,1)); ax.set_yticks(np.arange(-6,11,1))
        ax.grid(True, ls='--', lw=0.7)
        ax.set_title(f'f(x) = {a.value:.1f}·ReLU(x - {b.value:.1f}) + {c.value:.1f} | f(0)={y0:.2f}')
        ax.legend(loc='upper left', ncols=2)
        plt.show()

# dibuja y conecta eventos
dibujo()
for w in (a,b,c):
    w.observe(dibujo, names='value')

display(W.VBox([W.HBox([a,b,c]), out]))

````

### Problema 12 💻/🐍

A continuación, se presenta un gráfico dinámico de una función cuadrática, donde puedes variar los parámetros $a$, $b$ y $c$ para estudiar la forma de la función.

A partir de esto, explora el gráfico y responde las siguientes preguntas.

````python
#@title Parábola cuadrática con deslizadores (a, b, c)
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as W
from IPython.display import clear_output, display

# Eje x para graficar
x = np.linspace(-8, 8, 600)

# Sliders
a = W.FloatSlider(value=1.0, min=-3.0, max=3.0, step=0.1, description='a')
b = W.FloatSlider(value=0.0, min=-6.0, max=6.0, step=0.1, description='b')
c = W.FloatSlider(value=0.0, min=-6.0, max=6.0, step=0.1, description='c')
out = W.Output()

def draw(*_):
    with out:
        clear_output(wait=True)
        # función
        y = a.value*x**2 + b.value*x + c.value

        fig, ax = plt.subplots(figsize=(7, 4))
        ax.plot(x, y, lw=2, label='f(x) = a·x² + b·x + c')

        # Ejes y grilla
        ax.axhline(0, color='gray', lw=1)
        ax.axvline(0, color='gray', lw=1)
        ax.set_xlim(-8, 8); ax.set_ylim(-10, 10)
        ax.set_xticks(np.arange(-8, 9, 1))
        ax.set_yticks(np.arange(-10, 11, 1))
        ax.grid(True, which='both', linestyle='--', linewidth=0.7)

        # Corte con eje y (x=0): y = c
        ax.scatter([0], [c.value], s=60, color='black', label='intersección eje y')

        # Vértice y eje de simetría (si a != 0)
        eps = 1e-12
        if abs(a.value) > eps:
            xv = -b.value/(2*a.value)
            yv = a.value*xv**2 + b.value*xv + c.value
            ax.scatter([xv], [yv], s=70, color='red', zorder=5, label='vértice')
            ax.axvline(xv, color='purple', ls=':', lw=1.5, label='eje simetría')

            # Raíces reales si el discriminante es >= 0
            D = b.value**2 - 4*a.value*c.value
            if D >= 0:
                r1 = (-b.value - np.sqrt(D)) / (2*a.value)
                r2 = (-b.value + np.sqrt(D)) / (2*a.value)
                ax.scatter([r1, r2], [0, 0], s=70, color='green', label='raíces reales')
                roots_txt = f"raíces: {r1:.2f}, {r2:.2f}"
            else:
                roots_txt = "sin raíces reales"
            title_extra = f"  |  vértice=({xv:.2f}, {yv:.2f}), {roots_txt}"
        else:
            # Caso casi lineal: f(x) ≈ b x + c
            title_extra = "  |  (a≈0 ⇒ función casi lineal)"

        ax.set_title(f"f(x) = {a.value:.1f}·x² + {b.value:.1f}·x + {c.value:.1f}{title_extra}")
        ax.legend(loc='upper left', ncols=2)
        plt.show()

# dibuja y conecta
draw()
for w in (a, b, c):
    w.observe(draw, names='value')

display(W.VBox([W.HBox([a, b, c]), out]))
````
