## Herramientas nativas de Python

#### `print()`

La función print() recibe un string y lo muestra en pantalla.

Siempre muestra los string como formato de texto plano.

Antes del string se pueden agregar carácteres que definen la forma en que se lee el string.

El carácter f señala que es un string con formato, que permitirá insertar variables dentro del string.

Dentro del string, se pueden agregar términos como '\n' y '\t' que corresponden a saltos de línea y tabulaciones (4 espacios) respectivamente.

```python
a=5
b=10
print(f'El valor de a es\t{a}\n Y el valor de b es\t{b}')
```

Funciona de forma similar a print, sin embargo, no imprime sólo texto plano, sino que respeta los formatos propios de la librería, por tanto, será útil para expresiones matemáticas.

#### `range()`


Crea una iterador, a partir de tres argumentos, que en orden son el valor del comienzo, del fin y del paso.

- El comienzo está incluido y será el primer número de la lista.
- El fin no está incluído, de modo que se detendrá en el último número antes de él.
- El paso, determina cuántos valores se salta entre un término y otro.

```python
#Ejemplo desde 1 hasta 10, de 3 en 3
for i in range(1,10,3):
  print(i)
```

#### `round()`

Recibe como argumentos el número a redondear y la cantidad de decimales, si no se ingresa, es 0.

```python
a=1.674
print(a)          #Sin redondear
print(round(a,2)) #Dos decimales
print(round(a,1)) #Un decimal
print(round(a))   #Sin decimales
```

La librería Sympy permitirá el cálculo algebraico simbólico, de forma que resolverá los problemas de forma similar a cómo se resolvería "a mano".

Nos servirá para manipular funciones y expresiones, sin dar valores fijos.

#### Declaración de variables simbólicas

Para trabajar en sympy, necesitamos dar cuenta de las expresiones simbólicas con las que trabajaremos para diferenciarlas de las variables de Python.

Esto podemos hacerlo a partir de dos funciones, Symbol y symbols

```python
import sympy as sp

#Declara que la variable x ahora será el símbolo x en sympy
x=sp.Symbol('x')

#Declara que las variables a, b y c serán los símbolos a, b y c en sympy
a,b,c=sp.symbols('a b c')
```

#### Funciones elementales


Las usaremos para denotar funciones elementales (es decir, no polinómicas).

En este caso usaremos:
- exponencial
- logaritmo natural (en base $e$)
- logaritmos con base distinta a $e$

Es importante mencionar que, al definirlas de esta forma, son expresiones y no funciones, es decir, no se pueden evaluar.


```python
import sympy as sp
x=sp.Symbol('x')

print('exponencial')
f=sp.exp(x)
display(f)
print('\nlogaritmo natural')
g=sp.log(x)
display(g)
print('\nlogaritmo base 10')
h=sp.log(x,10)
display(h)
```

#### Creación de funciones

Para crear funciones, tendremos dos opciones.

Para funciones simbólicas, usaremos *sp.Lambda()*.

Esta función de *sympy* recibe dos argumentos, la variable independiente de la función, y la expresión matemática que caracteriza la función, de esta forma:

$$f(x)=x^2$$

Se escribe como:



```python
import sympy as sp
x=sp.Symbol('x')    #Crear variable simbólica
f=sp.Lambda(x,x**2) #Crear función simbólica
display(f)
display(f(2))
```

Este formato permite calcular valores directamente, pero no admite el uso sobre listas, que sí admite *numpy*.

En ese caso, podemos transformarlas a funciones compatibles con *numpy* a través de *sp.lambdify()*.

Esta función de sympy, recibe tres argumentos, la variable a usar, la expresión resultante, y el tipo de función que construirá, en este caso, numpy.

```python
import sympy as sp
import numpy as np
x=sp.Symbol('x')                  #Definimos la variable simbólica
x_num=np.arange(1,4,1)            #Definimos el rango a evaluar
f_num=sp.lambdify(x,x**2,'numpy') #Definimos el la función compatible con numpy
f_num(x_num)
```

Una forma alternativa, a partir de una función de tipo *sp.Lambda()* ya construída, es utilizar directamente las variables y la expresión de la función como *f.variables* y *f.expr* respectivamente.

```python
import sympy as sp
import numpy as np
x=sp.Symbol('x')                              #Creamos la variable simbólica
f=sp.Lambda(x,x**2)                           #Creamos la función simbólica

x_num=np.arange(1,4,1)                        #Indicamos los valores a evaluar
f_num=sp.lambdify(f.variables,f.expr,'numpy') #Creamos la función compatible con numpy
f_num(x_num)                                  #Evaluamos la función

```

O bien, lo mismo se puede escribir como:

```python
import sympy as sp
import numpy as np
x=sp.Symbol('x')                  #Creamos la variable Simbólica
f=sp.Lambda(x,x**2)               #Creamos la función simbólica

x_num=np.arange(1,4,1)            #Indicamos el rango de valores a evaluar, en este caso, [1,2,3]
f_num=sp.lambdify(x,f(x),'numpy') #Creamos la función compatible con numpy
f_num(x_num)                      #Evaluamos
```

#### Solución de Ecuaciones

Para resolver ecuaciones, Sympy tiene de forma analítica, usaremos en conjunto las herramientas sp.Eq() y sp.solve().

- sp.Eq() permite crear la ecuación que resolveremos, recibe dos argumentos, que corresponden a los lados izquierdo y derecho de la ecuación.

Así, la ecuación $x^2+2x=5$ se escribiría como:

```python
import sympy as sp
x=sp.Symbol('x')            #Definimos la variable simbólica
ecuacion=sp.Eq(x**2-5*x,6)  #Definimos la ecuación indicando ambos lados
display(ecuacion)
```

- sp.solve() nos permite resolver la ecuación, en torno a una variable. Para esto, indicaremos la ecuación a resolver y la variable que necesitamos despejar:



```python
import sympy as sp
x=sp.Symbol('x')              #Definimos la variable simbólica
ecuacion=sp.Eq(x**2-5*x,6)    #Definimos la ecuación, indicando lado izquierdo y derecho
solucion=sp.solve(ecuacion,x) #Resolvemos la ecuación
display(solucion)
```

Es importante notar que no todas las ecuaciones tienen solución analítica, es decir, que se pueden obtener despejando incognitas desde una ecuación, en los casos en que no se pueda, recurriremos a fsolve.

### Librería Scipy

#### `fsolve`

Desde la librería scipy en el módulo optimize usaremos la función `fsolve`.

Esta nos permitirá encontrar los ceros de una función de forma numérica. Para esto, requiere una como argumentos una función y una pista para comenzar a buscar en la cercanía.

Por tanto, en general será buena idea graficar primero para obtener una idea de dónde la función se anula y luego buscar.

Por ejemplo, para resolver la ecuación:

$x^2-5x=6$

Definimos la función auxiliar:

$f(x)=x^2-5x-6$

De esta forma, cuando $f(x)=0$ coincidirá con la solución de la ecuación:

```python
from scipy.optimize import fsolve

def f(x):
  return x**2-5*x-6       #Definimos la función para encontrar f(x)=0
pista=[0,5]               #Usualmente las obtendremos desde el gráfico
solucion=fsolve(f,pista)  #Se buscan las soluciones
display(solucion)
```

#### Límites de funciones


Los límites de funciones en *sympy* de pueden calcular mediante la función `sp.limit()`.

Esta función recibe necesariamente tres argumentos, que son, en orden, la expresión sobre la cual se aplica el límite, la variable respecto a la cual se aplica el límite, y el valor al cual se acerca.

De esta forma la expresión:

$$\lim_{x\to3}x^2$$

Se escribe como:

```python
import sympy as sp
sp.limit(x**2,x,3)
```

Por defecto, sympy calcula el límite por la derecha (es decir, bajando hacia el valor). Esto podemos modificarlo agregando el argumento dir de modo que:

```python
import sympy as sp
import numpy as np

print('Límite por la derecha:')
display(sp.limit(1/x,x,0,dir='+'))

print('Límite por la izquierda:')
display(sp.limit(1/x,x,0,dir='-'))

print('Límite bilateral:')
display(sp.limit(1/x,x,0,dir='+-'))
```

Aquí, la expresión: $$\tilde{\infty}$$ Representa que el límite diverge en magnitud, pero cambia de signo según la dirección del acercamiento y, por tanto, no existe.

#### Gráficos de sympy

Los gráficos de sympy son una herramienta que usaremos ocasionalmente, debido a que son más complejos de manipular que los de matplotlib, pero presentan algunas ventajas para funciones con divergencias en el dominio.

En este caso, asignaremos a la función `sp.plot()` los siguientes argumentos.
- Función que usaremos.
- La variable que usaremos y el rango de valores que puede tomar.
- `detect_poles=True`: Obliga al gráfico a respetar las divergencias, sin rellenarlas como haría matplotlib.
- `ylim(y_min,y_max)`: Determina el rango de valores del eje $y$ que se mostrará.
- `xlim(x_min,y_min)`: Determina el rango de valores del eje $x$ que se mostrará.

De modo que un ejemplo sería:

```python
import sympy as sp
import matplotlib.pyplot as plt

x = sp.Symbol('x')
f = sp.Lambda(x,2*x/(x**2 - 4))

sp.plot(f, (x, -5, 5), detect_poles=True, ylim=(-10, 10), xlim=(-5, 5))

plt.show()
```

### Librería Matplotlib

La librería matplotlib (en su módulo pyplot) permitirá graficar funciones dentro de un cierto rango y modificar la estética de ese gráfico.

Sin embargo, para funciones con discontinuidades tiende a rellenar los saltos.

Es por esto, que cuando tengamos funciones con discontinuidades, definiremos los dos rangos de forma independiente.

```python
#Importamos librerías
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

#Definimos la variable simbólica y la función adaptada
x=sp.Symbol('x')
f = sp.Piecewise(
    (x**2, x < 1),        # para x < 1
    (3, x >= 1)           # para x ≥ 1
)
f_num=sp.lambdify(x,f,'numpy')
#Definimos dos set de valores de x
x_num1=np.linspace(-5,0.99,100)               #Rango desde el inicio hasta antes de la discontinuidad
x_num2=np.linspace(1,5,100)                   #Rango posterior

#Definimos el espacio de trabajo
fig,ax=plt.subplots(figsize=(6,4))

#Construimos los gráficos
ax.plot(x_num1,f_num(x_num1),color='blue')    #Gráfico de líneas
ax.plot(x_num2,f_num(x_num2),color='blue')
#Títulos
ax.set_title('Función por parte')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')

#Agregamos los ejes del plano
ax.axhline(0,color='black',alpha=0.2)
ax.axvline(0,color='black',alpha=0.2)

#Personalziación
ax.grid()
ax.set_xlim(-5,5)
ax.ticklabel_format(style='plain', axis='both')

#Mostramos el gráfico.
plt.show()
```

#### Funciones elementales

Dentro de la librería sympy, encontramos diversas funciones elementales como lo son:

|Función elemental | Expresión matemática | Expresión en sympy|
|:--:|:--:|:--:|
|Raíz cuadrada| $\sqrt{x}$|`sp.sqrt(x)`|
|Raíz enésima | $\sqrt[n]{x}$|`sp.root(x,n)`|
|Logaritmo natural | $ln(x)$|`sp.log(x) o sp.ln(x)`|
|Logaritmo en base $a$|$log_a(x)$|`sp.log(x,a)`|
|Potencia | $x^n$|`x**n`|
|Exponencial en base a| $a^x$|`a**x`|
|Exponencial en base e| $e^x$|`sp.E**x`|
|Exponencial en base racional| $\bigg(\frac{a}{b}\bigg)^x$|`sp.Rational(a,b)**x`|

```python
import sympy as sp
x=sp.Symbol('x')
print('Raíz cuadrada')
display(sp.sqrt(x))
print('\nRaíz cúbica')
display(sp.root(x,3))
print('\nLogaritmo natural')
display(sp.ln(x))
print('\nLogaritmo natural')
display(sp.log(x))
print('\nLogaritmo base 2')
display(sp.log(x,2))
print('\nPotencia')
display(x**2)
print('\nExponencial')
display(2**x)
print('\nExponencial base e')
display(sp.E**x)
print('\nExponencial base racional')
display(sp.Rational(2,3)**x)
```

#### Simplificar expresiones

A veces, algunas expresiones complicadas pueden ser simplificadas, realizando divisiones u otras operaciones simples.

Generalmente sympy las realiza de forma automática, sin embargo, cuando esto no ocurre, se puede utilizar la función simplify sobre una expresión como:

`sp.simplify(expresion)`

```python
import sympy as sp

x=sp.Symbol('x')
expresion=(x**5-x**3)/x**2
print('Expresion sin simplificar:\n')
display(expresion)
print('\nExpresion simplificada:\n')
display(sp.simplify(expresion))
```

#### Función por parte

Para las funciones por parte usaremos `sp.Piecewise()`.

Esta expresión recibe una serie de duplas, con paréntesis, de una expresión y una condición, de modo que se evaluará en orden, cada una de las condiciones y la primera que sea cierta, será la expresión que se usará.

La expresión por defecto (símil a else) será poner `True` donde iría la condición. `True` se mostrará como `otherwise` lo que significa 'en otros casos'.

Es importante mencionar que esta es la forma correcta de declarar funciones por parte, ya que la secuencia if, elif y else no es compatible con sympy.

```python
import sympy as sp
x=sp.Symbol('x')
f=sp.Lambda(x,sp.Piecewise((-x,x<0),(x,True)))

display(f(x))
```

#### Derivadas en sympy

Para derivar expresiones en sympy usaremos:

`sp.diff(expresion, variable)`

Es importante mencionar que tanto el argumento, como la salida son expresiones, no funciones.

Por tanto, podemos hacerlo de las siguientes formas:

```python
import sympy as sp

x=sp.Symbol('x')

#Expresión a expresión, tanto y como dy se definen como expresiones
y=x**2
dy=sp.diff(y,x)
display(dy)

#Función a expresión, y se define como función, por ende ya no derivamos "y", sino "y(x)"
y=sp.Lambda(x,x**2)
dy=sp.diff(y(x),x)
display(dy)

#Expresión a función, se crea una función para la derivada
y=x**2
dy=sp.Lambda(x,sp.diff(y,x))
display(dy)

#Función a función
y=sp.Lambda(x,x**2)
dy=sp.Lambda(x,sp.diff(y(x),x))
display(dy)
```
