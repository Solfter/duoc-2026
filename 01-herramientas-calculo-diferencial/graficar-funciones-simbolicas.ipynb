```python
# 1.  Importación de librerias
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# 2. Definición simbolica y adaptativa númerica
x = sp.symbols('x')
g = (1/500)*x**5 - (23/400)*x**4 + 0.58*x**3 - 2.36*x**2 + 3.2*x
f = sp.lambdify(x, g, 'numpy')

# 3. Generación de valores númericos
# x_num = np.linspace(-5, 10, 100)
x_num = np.linspace(0, 11, 500)

# 4. Creación del espacio de trabajo
fig, ax = plt.subplots(figsize=(6,4))

# 5. Construcción del gráfico
ax.plot(x_num, f(x_num))

# 6. Titulos y etiquetas
ax.set_title('Función')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')

# 7. Agregar los ejes del plano
ax.axhline(0, color='black', lw=0.2)
ax.axvline(0, color='black', lw=0.2)

# 8. Personalizacion del gráfico
ax.grid()
#ax.set_xlim(-5, 10)
#ax.set_ylim(-5000, 25000)

# 9. Mostrar el gráfico
plt.show()
```
