## **Análisis de latencia en un clúster para modelos de lenguaje**

Una startup de IA está desplegando un clúster de servidores para procesar peticiones de un nuevo modelo de lenguaje. Para que el sistema sea eficiente, el equipo de ingeniería debe monitorear cómo el tiempo de respuesta (Latencia) se ve afectado por el crecimiento del clúster a medida que pasa el tiempo.

<br>

El número de nodos activos $n$ en el clúster (medido en unidades) depende del tiempo $t$ (medido en minutos desde el inicio de la carga de datos) según la función:

$$n(t)=0,5t^2+5t+200$$

Existen dos configuraciones posibles para manejar la latencia $L$ (medida en milisegundos) en función del número de nodos $n$:

**Arquitectura Alfa**:
    $$L_A​(n)=0,8n+40$$

**Arquitectura Beta**:
    $$L_B​(n)=\dfrac{120n+4000}{n+20}$$​

*Nota técnica: En el ámbito de la computación distribuida, un clúster se define como un conjunto de computadoras que funcionan de manera coordinada como un recurso único para el procesamiento de datos, mientras que un nodo representa cada unidad de procesamiento o servidor individual que integra dicha estructura.*
