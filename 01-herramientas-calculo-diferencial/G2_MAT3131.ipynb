```python
print(f'Limite por la derecha: {sp.limit(P(t), t, sp.oo, dir="+")}')
print(f'Limite por la izquierda: {sp.limit(P(t), t, -1*sp.oo, dir="-")}')
```

### Problema 18 💻/🐍

Considere el siguiente gráfico interactivo, donde se presenta la función:

$$
L(u; b)=
\begin{cases}
2,3 & u\le 100,\\[4pt]
2,5+0,01\,(u-100)^2 + b, & u>100.
\end{cases}
$$

*Nota: El parámetro $b$ es un número que puede modificarse con el deslizador del gráfico.*


```python
# @title Función por tramos con continuidad
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

a = 100.0
U = np.linspace(0, 200, 800)

def L_right(u, b):
    return 2.5 + 0.01*(u - a)**2 + b

def ver(b=-0.2):
    y = np.where(U <= a, 2.3, L_right(U, b))

    plt.figure(figsize=(7,4))
    # trazo por tramos para no “unir” el salto
    plt.plot(U[U<=a], y[U<=a], 'C0', lw=2)
    plt.plot(U[U>=a], y[U>=a], 'C0', lw=2)

    # marcas del salto en u=a
    plt.plot(a, 2.3, 'o', color='C0', zorder=5)                       # cerrado
    plt.plot(a, L_right(a, b), 'o', mfc='none', mec='C0', ms=9, zorder=5)  # abierto

    es_cont = abs(2.3 - (2.5 + b)) < 1e-12
    plt.title(f'L(100-)=2.30,  L(100+)={2.5+b:.2f}')
    plt.xlabel('u'); plt.ylabel('L(u)'); plt.xlim(0,200); plt.ylim(0,10)
    plt.grid(True, ls='--', lw=0.6, alpha=0.7)
    plt.show()

_ = interact(ver, b=FloatSlider(value=1, min=-1.0, max=1.0, step=0.02, description='b'))


```

a) A partir del gráfico, ¿para qué valor de $b$ la función es continua?
