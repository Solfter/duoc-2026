### Problema 3 💻/🐍

Una startup de inteligencia artificial lanza una plataforma web que usa modelos de lenguaje.

Durante los 20 primeros días, el número de usuarios registrados se puede aproximar por la función:

$$U(t)=500(1-e^{-0,1t})$$

Donde $t$ corresponde al tiempo en días desde el lanzamiento.

a) Identifique las variables dependiente e independiente.

b) Indique el dominio contextualizado de la función.

c) Genere e imprima una tabla que muestre la cantidad de usuarios registrados durante cada uno de los 10 primeros días desde el lanzamiento de la plataforma web.

```python
import sympy as sp

x = sp.Symbol('x')

f = sp.Lambda(x, 500*(1 - sp.exp(-0.1*x)))

display(f)

for i in range(20):
    print("="*30)
    resultado = f(i+1)
    print(f'Dia {i+1}: {int(resultado)} usuarios')
```

d) ¿Durante qué día se alcanzan los 400 usuarios?

```python
ecuacion=sp.Eq(500*(1 - sp.exp(-0.1*x)), 400)    #Definimos la ecuación, indicando lado izquierdo y derecho
solucion=sp.solve(ecuacion,x) #Resolvemos la ecuación
display(solucion)
```

e) Grafique la función $U(t)$ considerando el dominio contextualizado.

```python
#Importamos librerías
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

#Definimos la variable simbólica y la función adaptada
x=sp.Symbol('x')
f=sp.lambdify(x,500*(1 - sp.exp(-0.1*x)),'numpy')

#Definimos set de valores de x
x_num=np.linspace(0,20,100)

#Definimos el espacio de trabajo
fig,ax=plt.subplots(figsize=(6,4))

#Construimos los gráficos
ax.plot(x_num,f(x_num))    #Gráfico de líneas

#Títulos
ax.set_title('Cantidad de usuarios a través del tiempo')
ax.set_xlabel('Tiempo (días)')
ax.set_ylabel('Cantidad de Usuarios')

# Personalización
ax.grid()
ax.set_xlim(0, 20)

#Mostramos el gráfico.
plt.show()
```
