## **Recordatorio clave**

- **Estimador de la proporción:**$\hspace{0.5 cm}$  $\hat{p} = \frac{x}{n}$,
  donde $x$ es el número de éxitos y $n$ el tamaño de la muestra.

- **Error estándar:** $\hspace{0.5 cm}$ $SE = \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$

- **Intervalo de confianza para la proporción:** $\hspace{0.5 cm}$ $IC = \hat{p} \pm z_{\alpha/2} \cdot SE$

<br>
<br>

### **Comprobando condiciones antes de calcular IC**

Antes de aplicar las fórmulas anteriores, verifica estas 2 condiciones obligatorias en tus datos:

1.  **¿La muestra es aleatoria?**
    * Debe ser representativa del contexto del problema.
2.  **¿Hay suficientes éxitos y fracasos?** (Condición de Normalidad)
    * Calcula $n \cdot \hat{p}$ (debe ser $\geq 5$)
    * Calcula $n \cdot (1-\hat{p})$ (debe ser $\geq 5$)

*Si ambas respuestas son "SÍ", puedes proceder con el cálculo.*

# **Actividad 1 — Servidores estables en un centro de datos**

En un centro de datos que soporta múltiples aplicaciones críticas, la estabilidad de los servidores es esencial para garantizar disponibilidad y continuidad de servicio. Durante una semana de monitoreo se revisaron 200 servidores, registrando si experimentaron reinicios inesperados que podrían implicar caídas momentáneas o pérdida de datos. En total, se detectaron 26 servidores con incidentes de inestabilidad. El equipo de infraestructura necesita estimar cuál es la proporción real de servidores que podrían presentar este tipo de problemas para priorizar acciones de mantenimiento preventivo y asignar recursos de forma eficiente.

1. Calcula la proporción muestral $\hat{p}$ de servidores inestables.

```python
x = 26
n = 200
p_muestra = x/n

print(f"La proporción muestral es {p_muestra:.0%}")
```

2. ¿Es posible comprobar que es posible utilizar una aproximación normal?

```python
casos_exito = n * p_muestra >= 5 # Casos de exitos minimos
casos_fracaso = n * (1 - p_muestra) >= 5 # Casos de fracaso

if casos_exito & casos_fracaso:
    print("""
          Se cumplen las condiciones de éxito y fracaso mínimas.
          Entonces la muestra es válida para hablar de una aproximación normal.
    """)
else:
    print("No se cumplen las condiciones")
    
```

2. Construye un IC al 95% para la proporción poblacional.  


```python
alfa = 0.05
z = st.norm.ppf(1 - alfa/2)
SE = np.sqrt(p_muestra * (1 - p_muestra)/n)
IC = (p_muestra - z * SE, p_muestra + z * SE)
print(f"IC 95% = [ {IC[0]:.4f} ; {IC[1]:.4f} ]")
```

3. Interpreta el resultado para el equipo técnico.

**Basado en el IC, en el mejor de los casos el 8,34% de los servidores va a fallar, mientras que en el peor de los casos el 17,66% de los servidores va a fallar, al 95% de confianza**

# **Actividad 2 — Tasa de aprobación de un examen en educación**

En una asignatura clave de una carrera en educación superior, aprobar los exámenes iniciales es fundamental para medir el aprendizaje y orientar apoyos tempranos. En la primera evaluación participaron 120 estudiantes, de los cuales 72 aprobaron. El equipo docente quiere estimar la proporción real de estudiantes que aprueban este tipo de evaluaciones para tomar decisiones sobre reforzamientos y ajustar la planificación de la asignatura.

1. Calcula $ \hat{p}$ y construye un intervalo de confianza del 95% para la proporción poblacional. Imprime los resultados en representación decimal y porcentual.

```python
n = 120
x = 72
p_muestra = x/n

alfa = 0.05
z = st.norm.ppf(1 - alfa/2)
SE = np.sqrt(p_muestra * (1 - p_muestra)/n)
IC = (p_muestra - z * SE, p_muestra + z * SE)
print(f"IC 95% = [ {IC[0]:.4f} ; {IC[1]:.4f} ]")
print(f"IC 95% = [ {IC[0]:.2%} ; {IC[1]:.2%} ]")
```

2. Interpreta para el equipo docente: ¿cuál es el rango de aprobación estimado en todo el grupo de estudiantes?

**Podemos decir con un 95% de confianza, el interval de confianza poblacional se encuentra entre 51.23% y 68.77%**

**Si se dicta este curso y se vuelven a tomar muestras representativas, el 95% de las veces dará que el porcentaje de aprobación se encuentra entre 51.23% y 68.77%**

**En términos estadísticos para este tipo de asignaturas el peor de los casos la aprobación del curso será de 51.23% y en el mejor de los casos de 68.77%**

# **Actividad 3 — Satisfacción de clientes en compras online**

Una empresa de comercio electrónico está evaluando la calidad de su servicio al cliente. Para esto revisó 500 compras recientes, preguntando a los compradores si el servicio recibido fue “satisfactorio”. Entre ellos, 420 clientes respondieron afirmativamente. La empresa busca estimar, con un 90% de confianza, la proporción real de clientes satisfechos y comprender cómo cambiaría la precisión de la estimación si se aumentara el nivel de confianza a 99%.

1. ¿Cuál es el valor de la proporción muestral $ \hat{p}$?

```python
n = 500
x = 420
p_muestra = x/n
print(p_muestra)
```

2. Estima e interpreta el intervalo de confianza al 90% para la proporción poblacional de clientes satisfechos.

```python
alfa = 0.1

for alfa in [0.10, 0.01]:
    z = st.norm.ppf(1 - alfa/2)
    SE = np.sqrt(p_muestra * (1 - p_muestra) / n)
    IC = (p_muestra - z * SE, p_muestra + z * SE)
    print(f"IC {1 - alfa:.0%} = [ {IC[0]:.2%} ; {IC[1]:.2%} ]")
```

3. ¿Cómo cambia el tamaño del intervalo y la precisión de la estimación al aumentar el nivel de confianza del 90% al 99%?

**A mayor nivel de confianza, será más amplio el intervalo y menos precico en la estimacion**

## **Gráfico interactivo**

Utiliza el siguiente gráfico interactivo para explorar cómo cambia el intervalo de confianza al modificar el nivel de confianza, el tamaño muestral ($n$) y el valor de $x$.

Al mover los deslizadores, observa especialmente:
* Cómo varía el ancho del intervalo al cambiar el nivel de confianza,
* Qué ocurre con la precisión de la estimación.

*Indicación: Ejecuta la celda de código para visualizar el gráfico interactivo*


```python
# @title
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
from ipywidgets import interact,  IntSlider, Dropdown, fixed

def viz_ic_proporcion_curva(x, n, confianza=0.95, titulo=None):
    # Evitar casos inválidos
    if x < 0 or x > n:
        return

    phat = x / n
    z = st.norm.ppf(1 - (1-confianza)/2)
    se = np.sqrt(phat*(1 - phat)/n)
    L, U = phat - z*se, phat + z*se

    # Rango para la curva normal aproximada
    xs = np.linspace(max(0, phat - 4*se), min(1, phat + 4*se), 800)
    ys = st.norm.pdf((xs - phat)/se) / se

    # Gráfico
    plt.figure(figsize=(7,4))
    plt.plot(xs, ys)

    # sombreado del intervalo de confianza
    mask = (xs >= L) & (xs <= U)
    plt.fill_between(xs, ys, where=mask, alpha=0.3)
    plt.axvline(phat, linestyle="--", linewidth=1)
    plt.axvline(L, linestyle="--", linewidth=1)
    plt.axvline(U, linestyle="--", linewidth=1)

    if titulo is None:
        titulo = f"IC {int(confianza*100)}% para p con aproximación normal"
    plt.title(titulo)
    plt.xlabel("Proporción")
    plt.ylabel("Densidad (aprox. normal)")
    plt.text(phat, max(ys)*0.9, f"$\\hat p$={phat:.3f}", ha="center")
    plt.text(L, max(ys)*0.6, f"L={L:.4f}", ha="center", rotation=90, va="center")
    plt.text(U, max(ys)*0.6, f"U={U:.4f}", ha="center", rotation=90, va="center")
    plt.tight_layout()
    plt.show()

# --- INTERACTIVIDAD (único agregado real) ---
interact(
    viz_ic_proporcion_curva,
    x=IntSlider(value=70, min=0, max=500, step=10, description='x'),
    n=IntSlider(value=200, min=20, max=500, step=10, description='n'),
    confianza=Dropdown(options=[0.90, 0.95, 0.99], value=0.90, description='Confianza'),
    titulo=fixed(None)
);
```

4. ¿Qué representa el nivel de confianza en un IC y cómo se interpreta en el contexto de este problema?

El nivel de confianza representa el grado de seguridad asociado al método utilizado para construir el intervalo de confianza.

En este problema, decir que el intervalo tiene un nivel de confianza del 90% o del 99% significa que, si se repitiera el procedimiento muchas veces con distintas muestras del mismo tamaño, aproximadamente el 90% o el 99% de los intervalos construidos contendrían la verdadera proporción poblacional.
