### 1.Cargue los datos en un dataframe

```python
import kagglehub
import pandas as pd
import glob

# Descargar el dataset (sin extensión en el slug)
path = kagglehub.dataset_download("parulpandey/forbes-highest-paid-athletes-19902019")

# Encontrar el archivo CSV automáticamente
csv_files = glob.glob(f"{path}/**/*.csv", recursive=True)
print("Archivos encontrados:", csv_files)

# Cargar el primer CSV encontrado
df = pd.read_csv(csv_files[0])
df.head()
```

### 4. Transforme la variable “earnings” en una variable categórica implementando intervalos de clase. Calcule la frecuencia de cada clase.

```python
df['earning'], limites = pd.qcut(df['earnings ($ million)'], q=5, labels=['Muy Bajo', 'Bajo', 'Medio', 'Alto', 'Muy Alto'],retbins=True)
print(limites)
df['earning'].value_counts()
```

### 5. Conteste las siguientes preguntas:

#### a) ¿Quién es la persona que más dinero ganó en un año?

```python
df['Name'][df['earnings ($ million)'].idxmax()]
```

#### b) ¿Cuál es el deporte que más se repite dentro de la lista?

```python
deporte_moda = df['Sport'].mode()[0]
print(f"El deporte más común es: {deporte_moda}")
```

#### c) ¿Cuál es el deporte donde más dinero se gana?

```python
deporte_que_mas_gana_promedio = df.groupby('Sport')['earnings ($ million)'].mean().idxmax()
print(f"El deporte que más gana en promedio es: {deporte_que_mas_gana_promedio}")

deporte_que_mas_gana_total = df.groupby('Sport')['earnings ($ million)'].sum().idxmax()
print(f"El deporte que más gana en total es: {deporte_que_mas_gana_total}")
```

#### d) Si quisiera apostar a un jugador de Basketball según su ranking. ¿A cuál debería apostar?

```python
basket_df = df[df['Sport'] == 'Basketball']
mejor_promedio = basket_df.groupby('Name')['Current Rank'].mean().sort_values(ascending=True)
top_basketball = mejor_promedio.head(1)
print(mejor_promedio)
print(f"{top_basketball.index[0]} tiene el ranking más alto en basketball con el Ranking Promedio de {top_basketball.iloc[0]}")
```
